# User Age Distribution

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(9,8))
abc.Age.hist(facecolor='purple')
plt.title('User Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Users')
plt.show()
```

# User Rating of the Movie “Toy Story”

```python
plt.figure(figsize=(9,8))
d = abc.groupby('MovieName')
toy_story = d.get_group('Toy Story (1995)')
toy_story['Ratings'].hist(facecolor= 'forestgreen')
plt.title('User rating of thje movie “Toy Story”')
plt.xlabel('Ratings')
plt.ylabel('Number of Users')
plt.show()
```
