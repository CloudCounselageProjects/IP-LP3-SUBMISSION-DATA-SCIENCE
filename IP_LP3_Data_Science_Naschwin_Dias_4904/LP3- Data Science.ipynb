## Users Age Distribution

```python
age_dist = pd.DataFrame(users_df.groupby(['Age']).size(), columns = ['Count'])
age_dist
```

```python
range_list = users_df.groupby(['Age']).size().index.to_list()
rang = []
for i in range(len(range_list)-1):
    rang.append(f"{range_list[i]} - {range_list[i+1] - 1}")
rang.append('56+')
rang
```

```python
age_dist.index = rang
age_dist
```

```python
# Seaborn Barplot
fig, ax = plt.subplots(figsize=[10,8])
plt.rcParams['font.sans-serif'] = 'Arial'
plt.rcParams['font.family'] = 'sans-serif'
plt.rcParams['text.color'] = '#111'
plt.rcParams['xtick.color'] = '#101010'
plt.rcParams['ytick.color'] = '#101010'
plt.rcParams['font.size'] = 12
labels = age_dist.index
percentages = age_dist.Count
explode=(0.1,0)
ax.pie(percentages, labels=labels, startangle=0, pctdistance=1.2,labeldistance=1)
ax.axis('equal')
ax.set_title("Age Distribution")
ax.legend(frameon=False, bbox_to_anchor=(1.5,0.8))

#Seaborn barplot
sns.set_style("whitegrid")
bar,ax = plt.subplots(figsize=(10,8))
ax = sns.barplot(x=age_dist.index, y='Count', data=age_dist, ci=None, palette="muted",orient='v')
ax.set_title("Pie chart - Age Distribution", fontsize=15)
ax.set_xlabel ("Age Range")
ax.set_ylabel ("Number of People")
for rect in ax.patches:
    ax.text(rect.get_x() + rect.get_width() / 2,rect.get_height(),rect.get_height(), weight='bold', ha = 'center' )
```

## Using Column Genre to Separate them into Unique Column

```python
genres = movies_df['Genre'].to_list()
gen = set()
for genre in genres:
    for eac in genre.split('|'):
        gen.add(eac)

print(gen)
gen = list(sorted(gen))
print(gen)
```
